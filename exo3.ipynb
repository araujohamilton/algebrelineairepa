```python
£# nombre premier utilisé comme repere: 17 
# les restes possibles pour une division par 17: 0,1,...,16 
n = 17
from itertools import combinations
triples = list(combinations(range(n),3))
somme_triples = [sum(triples[i]) for i in range(len(triples))]
somme_triples_mod17 = [(i,somme_triples[i]%17) for i in range(len(triples))]
len(somme_triples_mod17)
```

```python
# Quels triples ont la somme = 0 module 17 ? combien de triples
l=[]
for i in somme_triples_mod17:
    if i[1]==0:
        l.append(i[0])
print(l, len(l))
```

```python
# Restriction du dataset aux cibles
triples_cibles = [triples[i] for i in l]
################ changer le index de a ######
a = triples_cibles[8]
#############################################
data = df[(df['mod17'] == a[0]) | (df['mod17'] == a[1]) | (df['mod17'] == a[2])]
```
