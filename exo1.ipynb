```python
import pandas as pd
import numpy as np
data = open('txt_exo1.txt')
```

```python
liste = [line for line in data]
conso_adresses = [liste[i][:-1] for i in range(len(liste))]
d = {'conso_adresses': conso_adresses}
df = pd.DataFrame(data=d)
df
```

```python
liste_conso = [int(df['conso_adresses'][i][0:df['conso_adresses'][i].index(' ')]) for i in range(1000)]
conso_max = max([int(df['conso_adresses'][i][0:df['conso_adresses'][i].index(' ')]) for i in range(1000)])
index_conso_max = liste_conso.index(conso_max)
df['conso_adresses'][index_conso_max]
```
